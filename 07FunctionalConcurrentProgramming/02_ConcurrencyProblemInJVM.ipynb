### Race Condition - 2 Threads are trying to set the same memory zone at the same time.These can introduce bugs in multithreaded code

```scala
object ThreadPoolDemo extends App{
    
    def runInParallel{
        var x=0
        val thread1=new Thread(new Runnable{
            override def run():Unit={x=1}
        })
        val thread2=new Thread(new Runnable{
            override def run():Unit={x=2}
        })
        thread1.start()
        thread2.start()
        Thread.sleep(8)
        println(x)        
    }
    
    for(i<-1 to 10) runInParallel
    
}
```

```scala
ThreadPoolDemo.main(Array())
```

```scala
object DemoTrans extends App{
    
    class BankAccount(var amount:Int){
        override def toString=""+amount
    }
    
    def buy(account:BankAccount,thing:String,price:Int){
        account.amount=account.amount-price
        println("I have bought : "+thing)
        println("My Account is now : "+account)
    }
    for(i<-1 to 1000){
        val acc=new BankAccount(50000)
        val thread1=new Thread(new Runnable{
            override def run():Unit=buy(acc,"Shoes",3000)
        })
        val thread2=new Thread(new Runnable{
            override def run():Unit=buy(acc,"Iphone",4000)
        })
        thread1.start()
        thread2.start()
        Thread.sleep(100)
        println("Current Amt: "+acc)
        println("----")
        
    }
    
}
```

```scala
DemoTrans.main(Array())
```

### Thread 1 and Thread 2 may both see the amount as 50000 and hence the balance will be 47000 for one 46000 for the other 

### Use Sychronised blocks  on references to avoid this race condition.NO two threads can evaluate the block at same time

```scala
object DemoTransWithSync extends App{
    
    class BankAccount(var amount:Int){
        override def toString=""+amount
    }
    
    def buy(account:BankAccount,thing:String,price:Int){
        account.synchronized{account.amount=account.amount-price
        println("I have bought : "+thing)
        println("My Account is now : "+account)
                            }
    }
    for(i<-1 to 1000){
        val acc=new BankAccount(50000)
        val thread1=new Thread(new Runnable{
            override def run():Unit=buy(acc,"Shoes",3000)
        })
        val thread2=new Thread(new Runnable{
            override def run():Unit=buy(acc,"Iphone",4000)
        })
        thread1.start()
        thread2.start()
        Thread.sleep(100)
        println("Current Amt: "+acc)
        println("----")
        
    }
    
}
```

### You can also use volatile against the variables that wil be modifed

```scala
object DemoTransWithVolatile extends App{
    
    class BankAccount(@volatile var amount:Int){
        override def toString=""+amount
    }
    
    def buy(account:BankAccount,thing:String,price:Int){
        account.amount=account.amount-price
        println("I have bought : "+thing)
        println("My Account is now : "+account)
    }
    for(i<-1 to 1000){
        val acc=new BankAccount(50000)
        val thread1=new Thread(new Runnable{
            override def run():Unit=buy(acc,"Shoes",3000)
        })
        val thread2=new Thread(new Runnable{
            override def run():Unit=buy(acc,"Iphone",4000)
        })
        thread1.start()
        thread2.start()
        Thread.sleep(100)
        println("Current Amt: "+acc)
        println("----")
        
    }
    
}
```

### Synchronise is more powerful as it allows you to put more expression in the same synchrnosed block
